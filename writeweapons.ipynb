```python
# Create navigation
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import time
# connecting and waiting for load
navegador.get("https://www.csgodatabase.com/skins/")
time.sleep(5)
```

```python
import pandas as pd
# reading the table
tabela = pd.read_excel("WeaponSkins.xlsx")
display(tabela)
```

```python
# finding the table
tbody = navegador.find_element(By.XPATH, '//*[@id="post-23"]/section/table[3]/tbody')
#finding the table rows
tr = tbody.find_elements(By.TAG_NAME, 'tr')
print(len(tr))
```

```python

for index, row in enumerate(tr, start=0):
  # taking the text from each cell in the table
  cells = row.find_elements(By.TAG_NAME, "td")
  cell_values = [cell.text for cell in cells]

  # Saving the values in the correct column
  tabela.loc[index, "Weapon"] = cell_values[0]
  tabela.loc[index, "Rarity"] = cell_values[1]
  tabela.loc[index, "Collection"] = cell_values[2]
  tabela.loc[index, "Launch Date"] = cell_values[3]
```

```python
# Cheking the table
display(tabela)
```

```python
# SAVING TABLE
tabela.to_excel("skinlinks.xlsx", index=False)
```

```python
# Cheking new table
links = pd.read_excel("skinlinks.xlsx")
display(links)
```

```python
# Inserting link column
links.insert(0, 'link', "")
```

```python
# copying weapons to use as our links
links["link"] = links["Weapon"]
display(links)
```

```python
# Cleaning especial characters
links["link"] = links["link"].str.replace(" ", "-", regex=True)
links["link"] = links["link"].str.replace("--", "-", regex=True)
links["link"] = links["link"].str.replace("_", "-", regex=True)
links["link"] = links["link"].str.replace("- ", "", regex=True)
links["link"] = links["link"].str.replace(r"[♥&!.'龍王&弐壱()]", '', regex=True)
links["link"] = links["link"].str.replace("ö", 'o', regex=True)

display(links)
```

```python
# Saving
links.to_excel("skinlinks.xlsx", index=False)
```
