```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
import time
url = "https://csgoskins.gg/items/"
navegador.get("https://csgoskins.gg")
time.sleep(3)
```

```python
import pandas as pd

tabela = pd.read_excel("skinprices.xlsx")
display(tabela.loc[:,['link']])
```

```python
getskin = tabela.loc[1,'link'].lower()
navegador.get(url+getskin)
print(len(tabela['link']))
```

```python
# get link on table
for i, row in tabela.iterrows():
  # get url
  getskin = row['link'].lower()
  navegador.get(url+getskin)
  # wait for website load
  time.sleep(3)
  
  # find the the prices path
  price_list = navegador.find_element(By.XPATH, '/html/body/main/div[2]/div[1]/div[1]/div')
  prices = price_list.find_elements(By.TAG_NAME, 'a')

  #take the prices values
  for index, row in enumerate(prices, start=1159):
    cells = row.find_elements(By.TAG_NAME, "span")
    cell_values = [cell.text for cell in cells]

    # verify empty values
    if cell_values:

      #verify and handle stattraks
      if len(cell_values) > 2:
        cell_values[0] += " " + cell_values[1]
        del cell_values[1]

      column_name = cell_values[0]
      price_value = cell_values[1]

      tabela.loc[i, column_name] = price_value
      print(cell_values)
```

```python
display(tabela)
tabela.to_excel("skinprices.xlsx", index=False)
```
